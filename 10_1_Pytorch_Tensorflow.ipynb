Below code demonstrates how to generate random numbers using different probability distributions in TensorFlow. The random number generator is initialized with a specific seed (`42`), ensuring reproducibility of the results.

### Code Breakdown:

1. **Uniform Distribution**:
   - The `uniform()` method generates random numbers from a **uniform distribution** in the range specified by `minval` and `maxval`. In this case, the generated random numbers are between `0` and `10`.
   - **Uniform Distribution** means every number within the specified range has an equal chance of being selected.

2. **Truncated Normal Distribution**:
   - The `truncated_normal()` method generates random numbers from a **truncated normal distribution**, where values are drawn from a normal distribution but any value more than 2 standard deviations away from the mean is discarded.
   - This is useful when you want to sample from a normal distribution but avoid extreme outliers.

3. **Poisson Distribution**:
   - The `poisson()` method generates random numbers following a **Poisson distribution**, which models the number of events occurring within a fixed interval of time or space. The `lam` parameter defines the rate (mean) of the distribution, here set to `3.0`.


```python
import tensorflow as tf

# Create a random number generator
random_1 = tf.random.Generator.from_seed(42)

# Generate random numbers from a uniform distribution
uniform_tensor = random_1.uniform(shape=(3, 2), minval=0, maxval=10)
print("Uniform Tensor:\n", uniform_tensor.numpy())

# Generate random numbers from a truncated normal distribution
truncated_normal_tensor = random_1.truncated_normal(shape=(3, 2))
print("\nTruncated Normal Tensor:\n", truncated_normal_tensor.numpy())

# Generate random numbers from a Poisson distribution
# poisson_tensor = random_1.poisson(shape=(3, 2), lam=3.0)
poisson_tensor = tf.random.poisson(shape=(3, 2), lam=3.0)
print("\nPoisson Tensor:\n", poisson_tensor.numpy())
```

Below code demonstrates how to generate random numbers from different probability distributions using **TensorFlow** and visualize them using **matplotlib**. Four different distributions are used: **Uniform**, **Truncated Normal**, **Poisson**, and **Gamma**.

### Code Breakdown:

1. **Random Number Generation**:
   - A **random number generator** (`random_1`) is created using a specific seed (`42`) for reproducibility.
   - The number of samples for each distribution is set to `10,000` for better visualization.

2. **Uniform Distribution**:
   - **`random_1.uniform()`** generates random numbers uniformly distributed between `0` and `10`.
   - The resulting tensor is converted into a NumPy array (`uniform_samples`).

3. **Truncated Normal Distribution**:
   - **`random_1.truncated_normal()`** generates random numbers from a truncated normal distribution. The values are drawn from a normal distribution but truncated to remove extreme outliers.
   - The resulting tensor is converted into a NumPy array (`truncated_normal_samples`).

4. **Poisson Distribution**:
   - **`tf.random.poisson()`** generates random numbers from a Poisson distribution with a mean (`lam`) of `3.0`. The resulting tensor is converted into a NumPy array (`poisson_samples`).

5. **Gamma Distribution**:
   - **`tf.random.gamma()`** generates random numbers from a Gamma distribution with shape parameters `alpha=2.0` and `beta=1.0`. The resulting tensor is converted into a NumPy array (`gamma_samples`).

6. **Plotting**:
   - **Matplotlib** is used to create histograms for each distribution in a 2x2 grid. The histograms show the frequency distribution of the generated random numbers for each type of distribution.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# !pip install matplotlib # Install matplotlib if not already installed

# Create a random number generator
random_1 = tf.random.Generator.from_seed(42)

# Number of samples for each distribution
num_samples = 10000

# --- Uniform Distribution ---
uniform_tensor = random_1.uniform(shape=(num_samples,), minval=0, maxval=10)
uniform_samples = uniform_tensor.numpy()

# --- Truncated Normal Distribution ---
truncated_normal_tensor = random_1.truncated_normal(shape=(num_samples,))
truncated_normal_samples = truncated_normal_tensor.numpy()

# --- Poisson Distribution (using tf.random.poisson) ---
poisson_tensor = tf.random.poisson(shape=(num_samples,), lam=3.0)
poisson_samples = poisson_tensor.numpy()

# --- Gamma Distribution (Example) ---
gamma_tensor = tf.random.gamma(shape=(num_samples,), alpha=2.0, beta=1.0) # Example using Gamma distribution
gamma_samples = gamma_tensor.numpy()

# --- Plotting ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Create a 2x2 grid of plots

# Uniform Distribution Plot
axes[0, 0].hist(uniform_samples, bins=50, density=True, alpha=0.7)
axes[0, 0].set_title("Uniform Distribution")
axes[0, 0].set_xlabel("Random Numbers")
axes[0, 0].set_ylabel("Frequency")

# Truncated Normal Distribution Plot
axes[0, 1].hist(truncated_normal_samples, bins=50, density=True, alpha=0.7)
axes[0, 1].set_title("Truncated Normal Distribution")
axes[0, 1].set_xlabel("Random Numbers")
axes[0, 1].set_ylabel("Frequency")

# Poisson Distribution Plot
axes[1, 0].hist(poisson_samples, bins=np.arange(0, max(poisson_samples) + 2, 1), density=True, alpha=0.7)
axes[1, 0].set_title("Poisson Distribution (tf.random.poisson)")
axes[1, 0].set_xlabel("Random Numbers")
axes[1, 0].set_ylabel("Frequency")

# Gamma Distribution Plot
axes[1, 1].hist(gamma_samples, bins=50, density=True, alpha=0.7)
axes[1, 1].set_title("Gamma Distribution (Example)")
axes[1, 1].set_xlabel("Random Numbers")
axes[1, 1].set_ylabel("Frequency")


plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

TensorFlow provides an easy way to convert **NumPy arrays** into **TensorFlow tensors**. This is useful when you have data in NumPy format and want to perform TensorFlow operations on it.

You can convert a NumPy array to a TensorFlow tensor using `tf.convert_to_tensor()` or `tf.constant()`. Both functions allow you to create a TensorFlow tensor from an existing NumPy array.

  - **`start=1`**: The array will start from `1`.
  - **`stop=25`**: The array will stop at `25` (but `25` itself is not included, so the last value is `24`).
  - **`step=1`**: The step size between each value is `1`.
- **`dtype=np.int32`**: The `dtype` argument specifies the data type of the array elements, which is set to `int32` (32-bit integer).


```python
numpy_A = np.arange(1, 25, dtype=np.int32)
numpy_A
```

### Matrix multiplication (is all you need)

[`torch.matmul()`](https://pytorch.org/docs/stable/generated/torch.matmul.html)

Two rules for matrix multiplication

1. The **inner dimensions** must match:
  * `(3, 2) @ (3, 2)` won't work
  * `(2, 3) @ (3, 2)` will work
  * `(3, 2) @ (2, 3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
 * `(2, 3) @ (3, 2)` -> `(2, 2)`
 * `(3, 2) @ (2, 3)` -> `(3, 3)`

> "`@`"  --  matrix multiplication.



```python
import torch
tensor = torch.tensor([1, 2, 3])
tensor.shape
```

```python
# Element-wise matrix multiplication
tensor * tensor
```

```python
# Matrix multiplication (@)
torch.matmul(tensor, tensor)
```

```python
tensor @ tensor
```

`torch.matmul()` method is faster

Neural networks are full of matrix multiplications and dot products.

The [`torch.nn.Linear()`](https://pytorch.org/docs/1.9.1/generated/torch.nn.Linear.html) module (we'll see this in action later on), also known as a feed-forward layer or fully connected layer, implements a matrix multiplication between an input `x` and a weights matrix `A`.

$$
y = x\cdot{A^T} + b
$$

Where:
* `x` is the input to the layer (deep learning is a stack of layers like `torch.nn.Linear()` and others on top of each other).
* `A` is the weights matrix created by the layer, this starts out as random numbers that get adjusted as a neural network learns to better represent patterns in the data (notice the "`T`", that's because the weights matrix gets transposed).
  * **Note:** You might also often see `W` or another letter like `X` used to showcase the weights matrix.
* `b` is the bias term used to slightly offset the weights and inputs.
* `y` is the output (a manipulation of the input in the hopes to discover patterns in it).

This is a linear function (you may have seen something like $y = mx+b$ in high school or elsewhere), and can be used to draw a straight line!

Let's play around with a linear layer.

Try changing the values of `in_features` and `out_features` below and see what happens.

Do you notice anything to do with the shapes?

```python
import torch

# Example tensor definition (adjust dimensions as per your requirement)
tensor_A = torch.randn(2, 3)  # Adjust the shape based on the expected input
```

```python
tensor_A = torch.randn(2, 2)  # Change the number of features to 2
```

```python
# Since the linear layer starts with a random weights matrix, let's make it reproducible (more on this later)
torch.manual_seed(42)
# This uses matrix multiplication
linear = torch.nn.Linear(in_features=2, # in_features = matches inner dimension of input
                         out_features=6) # out_features = describes outer value
x = tensor_A
output = linear(x)
print(f"Input shape: {x.shape}\n")
print(f"Output:\n{output}\n\nOutput shape: {output.shape}")
```

### Reshaping, stacking, squeezing and unsqueezing
---


| Method | One-line description |
| ----- | ----- |
| [`torch.reshape(input, shape)`](https://pytorch.org/docs/stable/generated/torch.reshape.html#torch.reshape) | Reshapes `input` to `shape` (if compatible), can also use `torch.Tensor.reshape()`. |
| [`Tensor.view(shape)`](https://pytorch.org/docs/stable/generated/torch.Tensor.view.html) | Returns a view of the original tensor in a different `shape` but shares the same data as the original tensor. |
| [`torch.stack(tensors, dim=0)`](https://pytorch.org/docs/1.9.1/generated/torch.stack.html) | Concatenates a sequence of `tensors` along a new dimension (`dim`), all `tensors` must be same size. |
| [`torch.squeeze(input)`](https://pytorch.org/docs/stable/generated/torch.squeeze.html) | Squeezes `input` to remove all the dimenions with value `1`. |
| [`torch.unsqueeze(input, dim)`](https://pytorch.org/docs/1.9.1/generated/torch.unsqueeze.html) | Returns `input` with a dimension value of `1` added at `dim`. |
| [`torch.permute(input, dims)`](https://pytorch.org/docs/stable/generated/torch.permute.html) | Returns a *view* of the original `input` with its dimensions permuted (rearranged) to `dims`. |


```python
import torch

x = torch.arange(1., 8.)
x, x.shape
```

```python

x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
```

`torch.view()`.

```python
z = x.view(1, 7)
z, z.shape
```

```python
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

 `torch.squeeze()`

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

`torch.unsqueeze()` to add a dimension value of 1 at a specific index.

```python
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```
